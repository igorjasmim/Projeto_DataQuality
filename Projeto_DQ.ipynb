```python
from dataquality import DataQuality
import pandas as pd 

# Read the DataFrame:
df = pd.read_csv('Forbes_Richest_Atheletes_1990_2020.csv')

# Iniciate the Class dataquality with the DataFrame:
dq = DataQuality(df)
```

## Check Methods:

### 1. Check Nulls

```python
# Check null values:
print("\nNull values for column:")
display(dq.check_nulls())
#dq.check_nulls()
#null_values = dq.check_nulls()
```

### 2. Check Duplicated

```python
# Check duplicated lines:
print(f"\nDuplicated lines: {dq.check_duplicates()}")
```

### 3. Check Numeric Columns

```python
# Check numeric columns:
print("\nNumeric columns:")
print(dq.check_numeric_cols())
```

### 4. Check Data Types

```python
# Check Data Types:
print("\nData Types:")
print(dq.check_data_types())
```

### 5. Check Unique Values

```python
# Check Unique Values:
print("\nUnique Values:")
print(dq.check_unique_values())
```

### 6. Describe Data

```python
# Check Describe Data:
print("\nDescribe Data:")
display(dq.describe_data())
```

### 7. Missing Values Summary

```python
# Missing Values Summary:
print("\nMissing Values:")
display(dq.missing_values_summary())
```
